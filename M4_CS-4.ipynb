```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("BigMartSalesData.csv", encoding='latin1')
```

```python
df_2011 = df[df['Year'] == 2011]

monthly_sales = df_2011.groupby('Month')['Amount'].sum()

plt.figure(figsize=(10, 5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='blue')
plt.title("Total Sales Per Month in 2011 (Line Plot)")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
bars = plt.bar(monthly_sales.index, monthly_sales.values, color='green')
plt.title("Total Sales Per Month in 2011 (Bar Chart)")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1, 13))
plt.grid(axis='y')
```

```python
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom', ha='center')

plt.show()

# 3. Plot Pie Chart for Sales by Country in 2011
country_sales_scaled = country_sales.head(10) / country_sales.head(10).sum() * 100  # Convert to percentages

plt.figure(figsize=(6, 6))
plt.pie(country_sales_scaled, labels=country_sales.head(10).index,
        autopct=lambda p: f'{p:.1f}%', startangle=90, shadow=True)

plt.title("Top 10 Country-wise Sales Distribution (2011)")
plt.show()
plt.savefig("plot.png", dpi=80)  # Lower DPI to control size
```

```python

invoice_amounts = df_2011.groupby('InvoiceNo')['Amount'].sum()
invoice_sample = invoice_amounts.sample(n=500, random_state=42)

# Plot the sampled data
plt.figure(figsize=(10, 5))
plt.scatter(range(len(invoice_sample)), invoice_sample.values, alpha=0.5, color='purple')
plt.title("Scatter Plot of Sampled Invoice Amounts")
plt.xlabel("Invoice Index")
plt.ylabel("Total Invoice Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```
