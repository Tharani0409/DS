```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

movies_df = pd.read_csv("HollywoodMovies.csv")
```

```python
#Task 1: Find the highest-rated movie in the “Quest” story type
quest_movies = movies_df[movies_df["Story"] == "Quest"]
highest_rated_quest = quest_movies.loc[quest_movies["RottenTomatoes"].idxmax()]
print("Highest-rated 'Quest' movie:")
print(highest_rated_quest[['Movie', 'RottenTomatoes']])
print()
```

```python
# Task 2: Genre with the most movie releases
most_common_genre = movies_df["Genre"].value_counts().idxmax()
print("Genre with the most movie releases:", most_common_genre)
print()
```

```python
#Task 3: Top five movies with the costliest budgets
top5_budget = movies_df.nlargest(5, 'Budget')
print("Top 5 costliest movies:")
print(top5_budget[['Movie', 'Budget']])
print()
```

```python
#Task 4: Plot Net Profitability vs Rotten Tomatoes rating
movies_df["NetProfitability"] = movies_df["WorldGross"] - movies_df["Budget"]
plt.figure(figsize=(10, 6))
plt.scatter(movies_df["RottenTomatoes"], movies_df["NetProfitability"])
plt.xlabel("Rotten Tomatoes Rating")
plt.ylabel("Net Profitability (WorldGross - Budget)")
plt.title("Net Profitability vs Rotten Tomatoes Rating")
plt.grid(True)
plt.tight_layout()
plt.savefig("net_profit_vs_rating.png")
plt.show()
```

```python
# Task 6: Series operations
# 6.1 String case and length
series1 = pd.Series(['Amit', 'Bob', 'Kate', 'A', 'b', np.nan, 'Car', 'dog', 'cat'])
print("Lower case:")
print(series1.str.lower())
print("\nUpper case:")
print(series1.str.upper())
print("\nLength:")
print(series1.str.len())
```

```python
# 6.2 Strip operations
series2 = pd.Series([' Atul', 'John ', ' jack ', 'Sam'])
print("Stripped (both):")
print(series2.str.strip())
print("\nLeft strip:")
print(series2.str.lstrip())
print("\nRight strip:")
print(series2.str.rstrip())
```

```python
# 6.3 String split and expand
series3 = pd.Series(['India_is_big', 'Population_is_huge', np.nan, 'Has_diverse_culture'])
print("Split:")
print(series3.str.split('_'))
print("\nAccess element [0]:")
print(series3.str.split('_').str[0])
print("\nExpand:")
print(series3.str.split('_', expand=True))
```

```python
# 6.4 Replace 'X' or 'dog' with 'XX-XX'
series4 = pd.Series(['A', 'B', 'C', 'AabX', 'BacX', '', np.nan, 'CABA', 'dog', 'cat'])
print("Replace X or dog:")
print(series4.replace(to_replace=r'(X|dog)', value='XX-XX', regex=True))
```

```python
# 6.5 Remove dollar signs
series5 = pd.Series(['12', '-$10', '$10,000'])
print("Remove dollars:")
print(series5.replace(r'[$,]', '', regex=True).astype(float))
```

```python
# 6.6 Reverse lowercase words
series6 = pd.Series(['India 1998', 'big country', np.nan])
print("Reverse lowercase words:")
print(series6.map(lambda x: ' '.join([word[::-1] if word.islower() else word for word in str(x).split()]) if pd.notnull(x) else x))
```

```python
# 6.7 Alphanumeric check
series7 = pd.Series(['1', '2', '1a', '2b', '2003c'])
print("Is alphanumeric:")
print(series7.map(lambda x: str(x).isalnum()))
```

```python
# 6.8 Contains 'A'
series8 = pd.Series(['1', '2', '1a', '2b', 'America', 'VietnAm','vietnam', '2003c'])
print("Contains 'A':")
print(series8.str.contains('A', case=False, na=False))
```

```python
# 6.9 a|b|c flags
series9 = pd.Series(['a', 'a|b', np.nan, 'a|c'])
print("a|b|c presence:")
print(pd.DataFrame({
    'a': series9.str.contains('a', na=False),
    'b': series9.str.contains('b', na=False),
    'c': series9.str.contains('c', na=False)
}).astype(int))
```

```python
# 6.10 Merge two dataframes
df1 = pd.DataFrame({'key': ['One', 'Two'], 'ltable': [1, 2]})
df2 = pd.DataFrame({'key': ['One', 'Two'], 'rtable': [4, 5]})
merged_df = pd.merge(df1, df2, on='key')
print("Merged DataFrame:")
print(merged_df)
```
