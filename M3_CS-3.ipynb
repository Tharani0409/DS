```python
import pandas as pd
import re

df = pd.read_csv('FairDealCustomerData.csv')
df.columns = df.columns.str.strip()
df = pd.read_csv('FairDealCustomerData.csv', names=['Name', 'Title', 'BlackListed'])
print(df.head())
```

```python
print(df.columns)
```

```python
#Define custom exception
class CustomerNotAllowedException(Exception):
    def __init__(self, message="Customer is blacklisted and not allowed to place an order"):
        self.message = message
        super().__init__(self.message)

#Define Customer class
class Customer:
    def __init__(self, name, title, blacklisted):
        self.title = title
        self.first_name, self.last_name = self.parse_name(name)
        self.blacklisted = blacklisted

    def parse_name(self, name):
        match = re.match(r'(\w+)\.?\s+(\w+)\s+(\w+)', name)
        if match:
            return match.groups()[1:]  # Skip the title since it's a separate field
        else:
            return ("", "")

    def __str__(self):
        return f"{self.title} {self.first_name} {self.last_name}, Blacklisted: {self.blacklisted}"

#Define Order class
class Order:
    def __init__(self, customer, product_name, product_code):
        self.customer = customer
        self.product_name = product_name
        self.product_code = product_code

    def __str__(self):
        return f"Order placed by {self.customer.first_name} {self.customer.last_name} for product {self.product_name} ({self.product_code})"

#Define createOrder function
def createOrder(customer, product_name, product_code):
    if customer.blacklisted == 1:
        raise CustomerNotAllowedException()
    return Order(customer, product_name, product_code)

#Test with sample records
for index, row in df.iterrows():
    cust = Customer(row['Name'], row['Title'], row['BlackListed'])
    try:
        order = createOrder(cust, "Smartphone", "SP123")
        print(order)
    except CustomerNotAllowedException as e:
        print(f"{cust.first_name} {cust.last_name}: {e}")
```
