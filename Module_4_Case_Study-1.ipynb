```python
#1.Extract data from the givenSalaryGender CSV file and store the data from each column in a separate NumPy array
import numpy as np
import pandas as pd 
df = pd.read_csv('SalaryGender.csv')
print(df)
```

```python
#2.Find:1. The number of men with a PhD. The number of women with a PhD
new_df = df.copy()
new_df['Gender'] = new_df['Gender'].replace([1,0], ['Male', 'Female'], inplace=False)
new_df['PhD'] = new_df['PhD'].replace([1,0], ['Yes', 'No'], inplace=False)
value = new_df[new_df['PhD']== 'Yes'].groupby(['Gender','PhD']).size().reset_index(name='count')
value
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#15.Data_Analysis
df = pd.read_csv('middle_tn_schools.csv')
df.head()
```

```python
df.info()
```

```python
df[['school_rating', 'size', 'reduced_lunch', 'full_time_teachers']] = df[['school_rating', 'size', 'reduced_lunch', 'full_time_teachers']].astype(int)
print(df.dtypes)
```

```python
df.describe().T
```

```python
reduced_lunch_stats = df.groupby('school_rating')['reduced_lunch'].describe()
print(reduced_lunch_stats)
```

```python
#chools with lower ratings (0–2) tend to have much higher percentages of students on reduced lunch programs indicating lower household incomes.
#Schools with higher ratings (4–5) tend to have lower percentages of students on reduced lunch suggesting a correlation between higher household income and better school ratings.
```

```python
y = df['reduced_lunch']
x = df['school_rating']
correlation = y.corr(x)
correlation 
```

```python
corr = df['school_rating'].corr(df['reduced_lunch'])
print(f"Correlation: {correlation:.4f}")

plt.figure(figsize=(10, 6))
sns.regplot(x='reduced_lunch', y='school_rating', data=df, scatter_kws={'alpha':0.6})
plt.title('Correlation between Reduced Lunch and School Rating')
plt.xlabel('Percentage of Students on Reduced Lunch')
plt.ylabel('School Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#The Pearson correlation coefficient is -0.8158 which is a strong negative correlation.
#This means that as the percentage of students receiving reduced lunch increases the school rating tends to decrease significantly.
```

```python
stu_teach_ratio_stats = df.groupby('school_rating')['stu_teach_ratio'].describe()
print(stu_teach_ratio_stats)
```

```python
b = df['stu_teach_ratio']
a = df['school_rating']
corr_stu = b.corr(a)
corr_stu 
```

```python
corr_graph = df['school_rating'].corr(df['stu_teach_ratio'])
print(f"Correlation: {corr_graph:.4f}")

plt.figure(figsize=(10, 6))
sns.regplot(x='stu_teach_ratio', y='school_rating', data=df, scatter_kws={'color': 'yellow'}, line_kws={'color': 'yellow'})
plt.title('Correlation between Student Teacher Ratio and School Rating')
plt.xlabel('Percentage of Student Teacher Ratio')
plt.ylabel('School Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#The scatter plot above shows the relationship between student-teacher ratio and school rating. The calculated correlation coefficient is approximately 0.2035 indicating a weak positive correlation.
#This means that in general schools with a higher student-teacher ratio tend to have slightly higher ratings though the relationship is not strong or definitive.
```

```python
corr_cols = ['school_rating', 'size', 'reduced_lunch', 'state_percentile_16','state_percentile_15', 'stu_teach_ratio', 'avg_score_15',
    'avg_score_16', 'full_time_teachers', 'percent_black','percent_white', 'percent_asian', 'percent_hispanic']

corr_mat = df[corr_cols].corr()

plt.figure(figsize=(16, 10))
sns.set(style="white")

sns.heatmap(corr_mat, annot=True, fmt=".2f", cmap="coolwarm", center=0,
            linewidths=0.5, cbar_kws={"shrink": 0.75})

plt.title("Correlation Matrix of School Dataset", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
School Rating has a strong negative correlation with Reduced Lunch (around -0.82), confirming that as the percentage of students on reduced lunch increases, school ratings tend to decrease.

School Rating also shows moderate positive correlations with:

State Percentile (15 & 16)

Average Scores (15 & 16)

Student-Teacher Ratio is negatively correlated with School Rating, but not as strongly.

Strong positive correlation exists between State Percentile and Average Scores — indicating these metrics reinforce each other.
```
