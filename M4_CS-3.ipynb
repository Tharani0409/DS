```python
import pandas as pd
import matplotlib.pyplot as plt

hurricane_df = pd.read_csv("Hurricanes.csv")

plt.figure(figsize=(15, 6))
plt.bar(hurricane_df['Year'], hurricane_df['Hurricanes'], color='skyblue')
plt.title('Number of Hurricanes by Year (Atlantic Coast, US)')
plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

city_temps_df = pd.read_csv("CityTemps.csv")

plt.figure(figsize=(10, 6))
plt.hist(city_temps_df['San Francisco'], bins=15, alpha=0.7, label='San Francisco', color='skyblue')
plt.hist(city_temps_df['Moscow'], bins=15, alpha=0.7, label='Moscow', color='salmon')
plt.title('Temperature Distribution (20142015)')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = [1, 2, 3, 4]
y = [20, 21, 20.5, 20.8]
y_error = [0.12, 0.13, 0.2, 0.1]

plt.figure(figsize=(4, 5), dpi=100)
plt.plot(X, y)
plt.title("Simple Line Plot", fontsize=14)
plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(4, 5), dpi=100)
plt.plot(X, y, marker='o', linestyle='--', color='green')
plt.title("Line with Markers", fontsize=14)
plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)
plt.grid(True)
plt.show()
```

```python
#Axes Configuration
plt.figure(figsize=(4, 5), dpi=100)
plt.plot(X, y, marker='s', color='blue')
plt.axis([0, 5, 19.5, 21.5])
plt.title("Configured Axes", fontsize=14)
plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)
plt.grid(True)
plt.show()
```

```python
#Error Bars
plt.figure(figsize=(4, 5), dpi=100)
plt.errorbar(X, y, yerr=y_error, fmt='-o', color='purple', ecolor='red', capsize=5)
plt.title("Error Bars Example", fontsize=14)
plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)
plt.grid(True)
plt.show()
```

```python
#Scatter Graph of 50 Random Points
x_rand = np.random.rand(50)
y_rand = np.random.rand(50)
plt.figure(figsize=(4, 5), dpi=100)
plt.scatter(x_rand, y_rand, color='orange')
plt.title("Random Scatter Plot", fontsize=14)
plt.xlabel("Random X", fontsize=14)
plt.ylabel("Random Y", fontsize=14)
plt.grid(True)
plt.show()
```
