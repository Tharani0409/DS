```python
import pandas as pd 
```

```python
# Load the data
df = pd.read_csv('CricketTestMatchData.csv')
```

```python
# Display the first few rows to check the data
df.head()
```

```python
# Rename columns for readability
df = df.rename(columns = {'Mat':'Matches', 'NO':'Not_Outs', 'HS':'Highest_Inns_Score', 'BF':'Balls_Faced', 'SR':'Batting_Strike_Rate'})
```

```python
df.head()
```

```python
# Check for missing values
df.isnull().any()
```

```python
df[df['Balls_Faced'].isna() == 1]
```

```python
# Handle missing values in 'Balls_Faced' and 'Batting_Strike_Rate'
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].fillna(0)
```

```python
df[df['Player'] == 'CL Walcott (WI)']
```

```python
# Check for duplicate rows
df.duplicated()
```

```python
df[df['Player'].duplicated() == 1]
```

```python
df[df['Player'].isin(['GA Headley (WI)', 'GS Sobers (WI)', 'JB Hobbs (ENG)', 'V Kohli (IND)'])]
```

```python
df = df.drop_duplicates()
```

```python
df[df['Player'].isin(['GA Headley (WI)', 'GS Sobers (WI)', 'JB Hobbs (ENG)', 'V Kohli (IND)'])]
```

```python
# Split 'Span' into 'Rookie_Year' and 'Final_Year'
df['Span'].str.split(pat = '-')
```

```python
df['Rookie_Year'] = df['Span'].str.split(pat = '-').str[0]
```

```python
df['Final_Year'] = df['Span'].str.split(pat = '-').str[1]
```

```python
df = df.drop(['Span'], axis = 1)
```

```python
# Extract 'Country' from 'Player'
df['Player'].str.split(pat = '(')
```

```python
df['Country'] = df['Player'].str.split(pat = '(').str[1]
```

```python
df['Country'] = df['Country'].str.split(pat = ')').str[0]
```

```python
df['Country']
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
```

```python
df['Player']
```

```python
df.head()
```

```python
# Handle '*' in 'Highest_Inns_Score' and convert to integer
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split(pat = '*').str[0]
```

```python
df['Highest_Inns_Score']
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype('int')
```

```python
df.dtypes
```

```python
# Convert 'Rookie_Year' and 'Final_Year' to integers
df = df.astype({'Rookie_Year':'int', 'Final_Year':'int'})
```

```python
df.dtypes
```

```python
# Clean 'Matches' column
df['Matches'] = df['Matches'].str.split(pat = '*').str[0]
```

```python
df['Matches'] = df['Matches'].astype('int')
```

```python
# Clean 'Balls_Faced' column
df['Balls_Faced'] = df['Balls_Faced'].str.split(pat = '+').str[0]
```

```python
df['Balls_Faced'] = df['Balls_Faced'].astype('int')
```

```python
df.isnull().any()
```

```python
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

```python
df[df['Balls_Faced'].isna()==1]
```

```python
df['Balls_Faced']
```

```python
# Clean '4s' and '6s' columns and convert to integers
df['4s'] = df['4s'].str.split(pat = '+').str[0]
```

```python
df['6s'] = df['6s'].str.split(pat = '+').str[0]
```

```python
df = df.astype({'4s':'int', '6s':'int'})
```

```python
df.dtypes
```

```python
# Calculate career length
df['Career_Length'] = df['Final_Year'] - df['Rookie_Year']
```

```python
df
```

```python
#Cricketers average career length 
df['Career_Length'].mean()
```

```python
#average batting strike rate for cricketers who played over 10 years 
df[df['Career_Length'] > 10]['Batting_Strike_Rate'].mean()
```

```python
#number of cricketers who played before 1960
df[df['Rookie_Year'] < 1960]['Player'].count()
```

```python
#max highest inns score by country 
df.groupby('Country')['Highest_Inns_Score'].max()
```

```python
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('HighInnsCountry').reset_index().sort_values('HighInnsCountry', ascending = False)
```

```python
#Hundreds, fifties, Ducks(0) avg by country
df.groupby('Country')[['100', '50', '0']].mean()
```
